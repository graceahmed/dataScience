## Abstract

The number of inmates held in U.S. State and Federal prisons skyrocketed from 294,400 in 1978, to 1,555,600 in 2009. This project attempts to visualize the changes in incarceration rate as well as examine correlates including crime, education, wealth, and the privatization of U.S. prisons in recent years.

## Data

The main data set used for this project was incarceration rates from 1978 to 2012 from the National Prisoner Statistics Program, downloaded from http://www.bjs.gov/nps/resources/documents/QT_imp%20rate_tot.xlsx as "data/QT_imp rate_tot(1).ods." I used Apache Open Office to crop the spreadsheet and save the pertinent portions as a csv file "data/state_imprionment_1978to2012.csv" and used Python's DictReader read the data and access it by year. After creating a function to access state incarceration rates by year, I mapped the data for all years 1978 through 2012 using shapefiles downloaded from the U.S. Census Bureau, https://www.census.gov/cgi-bin/geo/shapefiles2013/file-download. I spent many hours attempting to use Python's animation matplotlib.animation before fully understanding the nature of the package, and it's unsuitability to my goal. All maps were saved as png files in an images folder, and I used Boinx iStopMotion animation software to compile the still images into a movie file. 

To examine the relation between crime rates and incarceration rates, I downloaded downloaded FBI stats 1993-2012 as data/table_1_crime_in_the_united_states_by_volume_and_rate_per_100000_inhabitants_1993-2012.xls and saved the edited spreadsheet as data/usCrimeRate1993to2012.csv. From this csv, I created lists of violent crime rates and property crime rates for the years 1993-2012. I created a companion file of incarceration rates from the same period from data/QT_imp rate_tot(1).ods and saved it as data/usIncarcerationRate1993to2012.csv.

Data on education level by state was accessed from https://www.census.gov/compendia/statab/cats/education.html and downloaded as data/12s0233.xls. I saved 2009 data only as data/educationLevelsStates2009.csv

2009 Income level by state was accessed from http://www.census.gov/compendia/statab/2012/tables/12s0706.xls and downloaded as data/12s0706.xls. I saved states and median income columns only as data/medianIncomeStates2009.csv

Average net worth data for 2009 was cut and pasted from http://scorecard.assetsandopportunity.org/2012/measure/net-worth#fn-x. The source of the website's data was "Survey of Income and Program Participation, 2008 Panel, Wave 4. Washington, DC: U.S. Department of Commerce, Census Bureau, 2009. Data calculated by the Bay Area Council Economic Institute." This data set required more cleaning than the rest, as it contained html spacing characters, and had many missing values. For the 41 states net worth data was availible for, I matched incarceration rate by state.

Population density was copy & pasted from http://www.worldatlas.com/aatlas/populations/usadensityh.htm and saved as statePopDensityPerSqMile.csv

As I couldn't find a tidy and reliable source of information on states' use of private prisons, I decided to compare incarceration rate over time with stock prices for two major corporate owners of private prisions in the United States. I downloaded monthly stock prices for The GEO Group, Inc. (GEO) -NYSE July 28, 1994 through Dec 31, 2012 as data/GEOtable.csv from yahoo finace, and I downloaded monthly stock prices for Corrections Corporation of America (CXW) -NYSE  Jul 15, 1997 to Dec 31, 2012.

### Changes in U.S. Incarceration Rate 

```python
#http://stackoverflow.com/questions/16503560/read-specific-columns-from-csv-file-with-python-csv
columns = defaultdict(list) # each value in each column is appended to a list
with open('data/state_imprionment_1978to2012.csv') as f:
    reader = csv.DictReader(f, delimiter=';') # read rows into a dictionary format
    for row in reader: # read a row as {column1: value1, column2: value2,...}
        for (k,v) in row.items(): # go over each column name and value 
            columns[k].append(v) # append the value into the appropriate list
                                 # based on column name k
                             
#function to access data by year
#e.g. year_2000 = incarceration4year('2000')
def incarceration4year( year ):
   "takes a year as string and returns a dict with key:value pairs of state:incarceration rate"
   d = { columns['State'][i] : int(columns[year][i]) for i in range(0, 50)}
   return d
```

```python
#create color coded map of incarceration rates in lower 48 staes, D.C. excluded
#adapted from class lecture 2/27/14 http://bagrow.com/dsv/LEC14_notes_2014-02-27.html

# Lambert Conformal map of lower 48 states.
m = Basemap(llcrnrlon=-119,llcrnrlat=22,
            urcrnrlon=-64, urcrnrlat=49,
            projection='lcc', lat_1=33,lat_2=45,lon_0=-95)

# load state boundaries.
# data from U.S Census Bureau
# https://www.census.gov/cgi-bin/geo/shapefiles2013/file-download
shp_info = m.readshapefile('data/tl_2013_us_state/tl_2013_us_state','states',drawbounds=False)

def mapIncarcerationRate(year, rateByYear):
    #clear current figure to get ready to write
    plt.clf()
    
    # choose a color for each state based on incarceration rate.
    colors={}
    statenames=[]
    cmap = plt.cm.YlOrRd # use yellow orange red colormap
    vmin = 0; vmax = 900 # set range.
    sm = plt.cm.ScalarMappable(cmap=cmap, 
                               norm=plt.Normalize(vmin=vmin, vmax=vmax))
    
    for shapedict in m.states_info:
        statename = shapedict['NAME']
    
        try:
            inc_rate = float(rateByYear[statename])
        except KeyError:
            inc_rate = 0.0
            
        # calling colormap with value between 0 and 1 returns
        # rgba value. Invert color range (hot colors are high
        # population), take sqrt root to spread out colors more.
        colors[statename] = cmap((inc_rate-vmin)/(vmax-vmin))[:3]
        statenames.append(statename)
            
    # cycle through state names, color each one.
    for nshape,seg in enumerate(m.states):
        xx,yy = zip(*seg)
        if statenames[nshape] != 'District of Columbia' and \
           statenames[nshape] != "Puerto Rico":
            color = rgb2hex(colors[statenames[nshape]]) 
            plt.fill(xx,yy,color,edgecolor=color)
    
    # draw meridians and parallels.
    m.drawparallels(np.arange(25,65,20),   labels=[0,0,0,0],
                    zorder=-1,color="w")
    m.drawmeridians(np.arange(-120,-40,20),labels=[0,0,0,0],
                    zorder=-1,color="w")
    
    # set up colorbar:
    mm = plt.cm.ScalarMappable(cmap=cmap)
    mm.set_array([vmin,vmax])
    plt.colorbar(mm, label="Incarceration Rate per 100,000",
                 ticks=[0,150,300,450,600,750,900],
                 orientation="horizontal", fraction=0.05,
                )
    
    plt.title(year, fontsize=18)
    plt.gca().axis("off")
    
#create maps of incarceration rates for years 1978 through 2010 and save in images folder  
years = range(1978,2011)   
for year in years:
    rateByYear = incarceration4year(str(year))
    mapIncarcerationRate(year, rateByYear)
    savefig('images/ir'+str(year)+'.png')
    
    
#display maps of incarceration rates every 10 years
years = ['1980','1990','2000', '2010']
for year in years:
    rateByYear = incarceration4year(year)
    mapIncarcerationRate(year, rateByYear)
    plt.show()
```

```python
#peak us incarcercation rate 2008
#lowest in 1978

data1978 = incarceration4year('1978')
stateSortedIncarceration1978 = sorted(data1978.items(), key=lambda x: x[0])
incarcRate1978 = [int(i[1]) for i in stateSortedIncarceration1978]
data2008 = incarceration4year('2008')
stateSortedIncarceration2008 = sorted(data2008.items(), key=lambda x: x[0])
incarcRate2008 = [int(i[1]) for i in stateSortedIncarceration2008]
#the change in incarceration rate
deltaIncarcRate1978to2008 = []
for i in range(len(incarcRate2008)):
    deltaRate = incarcRate2008[i] - incarcRate1978[i]
    deltaIncarcRate1978to2008.append(deltaRate)
    
# choose a color for each state based on incarceration rate.
colors={}
statenames=[]
cmap = plt.cm.Greys # use Greys colormap for visual clarity
vmin = 0; vmax = 500 # set range.
sm = plt.cm.ScalarMappable(cmap=cmap, 
                           norm=plt.Normalize(vmin=vmin, vmax=vmax))

for shapedict in m.states_info:
    statename = shapedict['NAME']

    try:
        deltaInc_rate = float(data2008[statename]-data1978[statename])
    except KeyError:
        deltaInc_rate = 0.0
        
    # calling colormap with value between 0 and 1 returns
    # rgba value. Invert color range (hot colors are high
    # population), take sqrt root to spread out colors more.
    colors[statename] = cmap((deltaInc_rate-vmin)/(vmax-vmin))[:3]
    statenames.append(statename)
        
# cycle through state names, color each one.
for nshape,seg in enumerate(m.states):
    xx,yy = zip(*seg)
    if statenames[nshape] != 'District of Columbia' and \
       statenames[nshape] != "Puerto Rico":
        color = rgb2hex(colors[statenames[nshape]]) 
        plt.fill(xx,yy,color,edgecolor=color)

# draw meridians and parallels.
m.drawparallels(np.arange(25,65,20),   labels=[0,0,0,0],
                zorder=-1,color="w")
m.drawmeridians(np.arange(-120,-40,20),labels=[0,0,0,0],
                zorder=-1,color="w")

# set up colorbar:
mm = plt.cm.ScalarMappable(cmap=cmap)
mm.set_array([vmin,vmax])
plt.colorbar(mm, label="Rate per 100,000",
             ticks=[0,100,200,300,400,500],
             orientation="horizontal", fraction=0.05,
            )

plt.title('Increase in Incarceration Rate 1978 to 2008', fontsize=14)
plt.gca().axis("off")
plt.show()
```

### Comparing U.S. Incarceration Rate and U.S. Crime Rate: Does jailing criminals make us safer?

```python
#for analysis of violent crime / property crime / incarceration during this period

#load U.S. total incarceration rate per 100,000 population for years 1993-2012
    #http://stackoverflow.com/questions/2953746/python-parse-comma-separated-number-into-int
    #http://stackoverflow.com/questions/6740918/creating-a-dictionary-from-a-csv-file
with open('data/usIncarcerationRate1993to2012.csv', 'r') as csvfile:
    r = csv.reader(csvfile, delimiter=';')
    #create dictionary of years (as strings) to incarceration rate per 100,000 (as integers)
    dictIncarcRate1993to2012 = dict((rows[0].replace(',',''),int(rows[1])) for rows in r)

incarceration1993to2012 = []
for k in range(1993,2013):
    val = dictIncarcRate1993to2012.get(str(k))
    incarceration1993to2012.append(val)
    
#load U.S. crime rates per 100,000 population
vcr1993to2012 = []
pcr1993to2012 = []
with open('data/usCrimeRate1993to2012.csv', 'r') as f:
    r = csv.reader(f, delimiter=';')
    
    for rows in r:
        vcr1993to2012.append(rows[3]) 
        pcr1993to2012.append(rows[13])

violentCrimeRate1993to2012 =[]
for item in vcr1993to2012[1:]:
    violentCrimeRate1993to2012.append(float(item))

propertyCrimeRate1993to2012 = []
for item in pcr1993to2012[1:]:
    propertyCrimeRate1993to2012.append(float(item.replace(',','')))
```

```python
crimeTypeCorr = numpy.corrcoef(
    violentCrimeRate1993to2012,
    propertyCrimeRate1993to2012)[0, 1]
print "Correlation coefficent of violent crime rate 1993-2012 and property crime rate 1993-2012 is " + str(crimeTypeCorr)

incarcToViolentCrimeCorr = numpy.corrcoef(
    incarceration1993to2012,
    violentCrimeRate1993to2012)[0, 1]
print "Correlation coefficent of violent crime rate 1993-2012 and incarceration rate 1993-2012 is " + str(incarcToViolentCrimeCorr)

incarcToPropertyCrimeCorr = numpy.corrcoef(
    incarceration1993to2012,
    propertyCrimeRate1993to2012)[0, 1]
print "Correlation coefficent of property crime rate 1993-2012 and incarceration rate 1993-2012 is " + str(incarcToPropertyCrimeCorr)
```

### U.S. Incarceration Rate and Stock Prices for Private Prison Corporations 

```python
#get stock prices over time for CWX
with open('data/CXWtable.csv') as f:
    reader = csv.DictReader(f, delimiter=',') # read rows into a dictionary format
    cwxDays = []
    cwxPrices = []
    for row in reader:
        day = row.get('Date')
        day = time.strptime(day, "%Y-%m-%d")
        day = datetime.fromtimestamp(mktime(day))
        cwxDays.append(day)
        price = row.get('Adj Close')
        cwxPrices.append(price)

#get stock prices for GEO Group
with open('data/GEOtable.csv') as f:
    reader = csv.DictReader(f, delimiter=',') # read rows into a dictionary format
    geoDays = []
    geoPrices = []
    for row in reader:
        day = row.get('Date')
        day = time.strptime(day, "%Y-%m-%d")
        day = datetime.fromtimestamp(mktime(day))
        geoDays.append(day)
        price = row.get('Adj Close')
        geoPrices.append(price)        

#get data for total U.S. Incarceration Rate for years 1978-2010
with open('data/usIncarcerationTotal1978to2012.csv') as f:
    reader = csv.DictReader(f, delimiter=';') # read rows into a dictionary format
    incarcYears = []
    incarcRateTotal = []
    for row in reader:
        yr = row.get('Year')
        yr = yr.replace(',', '')
        yr = time.strptime(yr, "%Y")
        yr = datetime.fromtimestamp(mktime(yr))
        incarcYears.append(yr)
        incarc = row.get('U.S. Total incarceration')
        incarcRateTotal.append(incarc) 
```

In 2009, the states with the highest incarceration rates were Louisiana, Mississippi, Oklahoma, Texas, Alabama, and Arizona. The highest incarceration rate, 886 per 100,000 in Louisiana, was nearly 6 times as high as that of the lowest incarceration rate, 150 per 100,000 in Maine.

The data for median net worth is somewhat unreliable, as only 41 states were examined, and the source of estimation was not highly accurate. However, taken together, there appears to be a moderate negative correlation between wealth and incarceration rate.

Given the high p-value of the pearsons test, no correlation between state population density and state incarceration rate appears to exist.
